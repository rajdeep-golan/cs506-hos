```python
# automated_web.py

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from pyvirtualdisplay import Display
import time

# Start virtual display
display = Display(visible=0, size=(1024, 768))
display.start()

options = webdriver.ChromeOptions()
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--headless')  # No GUI mode
options.add_argument('--user-data-dir=/tmp/unique-user-data')  # avoid conflicts
options.binary_location = "/usr/bin/chromium-browser"

browser = webdriver.Chrome(options=options)

# Start automation
browser.get("https://accounts.google.com/signup")
time.sleep(2)

# Fill the form
browser.find_element(By.ID, 'firstName').send_keys('Thomas')
browser.find_element(By.ID, 'lastName').send_keys('Anderson')
browser.save_screenshot('step1_name_filled.png')
time.sleep(1)

browser.find_element(By.ID, "collectNameNext").click()
time.sleep(2)

browser.find_element(By.NAME, "day").send_keys('1')
Select(browser.find_element(By.ID, "month")).select_by_visible_text('January')
browser.find_element(By.NAME, "year").send_keys("1990")
Select(browser.find_element(By.ID, "gender")).select_by_visible_text("Male")
browser.save_screenshot('step2_birth_gender.png')
time.sleep(1)

browser.find_element(By.ID, "birthdaygenderNext").click()
time.sleep(2)

browser.find_element(By.NAME, "Username").send_keys("thomasanderson111990")
browser.save_screenshot('step3_username.png')
time.sleep(1)

browser.find_element(By.ID, "next").click()

passwd = WebDriverWait(browser, 60).until(EC.visibility_of_element_located((By.NAME, "Passwd")))
passwd.send_keys("ThomAnd1190")
passwdAgain = WebDriverWait(browser, 10).until(EC.visibility_of_element_located((By.NAME, "PasswdAgain")))
passwdAgain.send_keys("ThomAnd1190")

browser.save_screenshot('step4_password.png')
time.sleep(1)

browser.find_element(By.ID, "createpasswordNext").click()
time.sleep(2)

browser.save_screenshot('final_step.png')

browser.quit()
display.stop()
```

```python
# read_excel.py

import openpyxl

wb = openpyxl.load_workbook('/workspaces/cs506-hos03-spring-2025-Ashwin9515/Module03 - Ashwin Paturi/example.xlsx')
print(wb.sheetnames)
# wb.active selects the first available sheet (sheet 1)
sheet = wb.active
print(sheet.title)
for i in range(1, sheet.max_row+1):
    # print all data in column A,B,C
    # ljust() is used to make it easier to read
    print(i, str(sheet.cell(row=i, column=1).value).ljust(25),
          sheet.cell(row=i, column=2).value.ljust(15, ' '),
          str(sheet.cell(row=i, column=3).value))
```

```python
#write_excel.py

import openpyxl

wb = openpyxl.Workbook()
sheet = wb.active
sheet.title = "Cars"
sheet['A1'] = "Models"
sheet['B1'] = "Price"

cars = [('BMW', 40000), ('Audi', 50000), ('Ford', 25000), ('McLaren', 1800000), ('Toyota', 30000)]

for car in cars:
    sheet.append(car)
    # appends a group of values at the bottom of the current sheet
    # in order from first column

wb.save('car_data.xlsx')
```
