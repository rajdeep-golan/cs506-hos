```python
# dict_for.py

alien_0 = {'color': 'green', 'points': 5}
alien_1 = {'color': 'yellow', 'points': 10}

aliens = [alien_0, alien_1]

for i in aliens:
    for key, value in i.items():
        print("KEY: ", key, "\t" , "VALUE :", value)
```

```python
#dictionary.py

dict = {'Name': 'abc', 'Age': 7}
print("Name : ", dict['Name'] , "\n", "Age :", dict['Age'])

dict['Age']=20
print("Updated Age :", dict['Age'])

dict['Phone_no']=123456789
print("After adding the new pair :", dict)

del dict['Phone_no']
print("After deleting phone_no : ", dict)
```

```python
#ifControl.py

print("How old are you?")
age = int(input())

if age < 22:
    print("You are too young to have a drink.")

elif age >=80:
    print("Ok, you will get a free drink.")

else:
    print('Sure, enjoy your drink.')
```

```python
#lowest.py

income = {'Alice': 90000,
          'Bob': 100000,
          'Jeff': 200000,
          'Apiwat': 999998,
          'Stark': 999999

}

lowest = min(income, key=income.get)
print("The lowest income is :", end=' ')
print(lowest + ' with $' + str(income[lowest]))
```

```python
#replaceNegative.py

original = [8,20, -10, 55, -777]

for i in range(len(original)):
    original[i]=abs(original[i])

print(original)
```

```python
#richest.py

income = {'Alice': 90000,
          'Bob': 100000,
          'Jeff': 200000,
          'Apiwat': 999998,
          'Stark': 999999

}

highest = max(income, key=income.get)
print("The richest man on earth:", end=' ')
print(highest + ' with $' + str(income[highest]))
```
